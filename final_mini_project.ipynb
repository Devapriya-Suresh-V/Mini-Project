```
import os
from skimage.io import imread
from skimage.transform import resize
```

```
target = []
flat_data = []
images = []

path = '/content/drive/My Drive/Colab Notebooks/images/'
for dirpath, dirnames, filenames in os.walk(path):
  for filename in filenames:
    category = dirpath.split('/')[-1]
    target.append(category)
    img = imread(f'{path}{category}/{filename}')
    img_resized = resize(img,(50,50,50))
    flat_data.append(img_resized.flatten())
    images.append(img_resized)

flat_data = np.array(flat_data)
target = np.array(target)
images = np.array(images)

print (flat_data.shape)
print (target.shape)
print (images.shape)
```

```
import pandas as pd
df = pd.DataFrame(flat_data)
df['target'] = target
```

```
df.head(10)
```

```
print(df['target'])
```

```
y = df["target"].values
```

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```
x = df.iloc[:,0:125000].values
```

```
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 0)
```

```
x_train.shape
```

```
y_train
```

```
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import GridSearchCV
```

```
parameters = {
    'svc':{
        'model':SVC(),
        'params':{
            'C':[1, 10, 20],
            'kernel':['linear','rbf']
        }
    },
    'random_forest':{
        'model':RandomForestClassifier(),
        'params':{
            'n_estimators':[10,20,50,70,100]
        }
    },
    'decision_tree':{
        'model':DecisionTreeClassifier(),
        'params':{}
    },
    "naive_bayes":{
        'model':GaussianNB(),
        'params':{}
    }
}
```

```
scores = []

for model_name, mp in parameters.items():
    clf = GridSearchCV(mp['model'], mp['params'],cv = 5)
    clf.fit(x_train, y_train)
    scores.append({
        'model':model_name,
        'best_score':clf.best_score_,
        'best_params':clf.best_params_
    })
```

```
scores
```

```
classi = SVC(C=1,kernel='rbf')
```

```
classi.fit(x_train, y_train)
```

```
y_pred = classi.predict(x_test)
```

```
from sklearn.metrics import accuracy_score, confusion_matrix
```

```
ac = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
```
